```python
# calculadora de IMC
# formula IMC = Peso/(Altura ^ 2)
# Menor que 22 - Baixo peso
# De 22 a 27 - Peso normal
# de 27 a 29,99 - Sobrepeso
# Maior que 30 - Obesidade
# string = texto
# int = mostrar numero inteiro
# float = mostrar numero com casas decimais
# para representar potencia é necessario colocar dois asteriscos = ** 2
# round - serve para arredondar um numero = round(o calculo),'qnts num dps da ,'


print('# # # SEJA BEM VINDO A CALCULADORA DE IMC # # #')
nome = input('Qual o seu nome: ')
peso = float(input('Qual o seu peso: '))
altura = float(input('Qual a sua altura: '))

imc = round(peso /(altura ** 2 ),2)


if imc < 22:
   print(f'Olá {nome}, seu imc é {imc} e voce está abaixo do peso esperado, procure um nutricionista.')


```
